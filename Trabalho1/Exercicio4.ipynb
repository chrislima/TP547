### 4) Se ocorrerem falhas de energia elétrica de acordo com uma distribuição de Poisson com uma média de 6 falhas a cada duas semanas, calcule a probabilidade de que haverá ao menos 2 falhas durante uma semana específica. Traçar o histograma da variável analisada.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
lambda1 = 3
N = 100000
value = 2
count = 0
av = np.array([])
x = np.random.uniform(0,1,N)
for ix in x:
    i = 0
    pr = np.exp(-lambda1)
    F=pr
    while ix >= F:
        pr = lambda1/(i+1)*pr
        F += pr
        i += 1
    a1=i
    av=np.append(av,a1)
```

```python
for poivalue in av:
    if poivalue >= value:
        count += 1
prob=count/N
print("A probabilidade de haver pelo menos duas falhas P(X >= 2) = {:.2f}%".format(prob*100))
```

```python
plt.hist(av, bins=range(int(np.max(av)) + 2), density=True, alpha=0.7, color='blue')
plt.xlabel('Valores')
plt.ylabel('Frequência')
plt.title('Histograma ')
plt.show()
```
