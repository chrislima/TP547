# 7)  Utilizando o método da inversa gerar amostras para a distribuição f(x)=ex/((e2-1)) 0<=x<=2

### Pelo método da inversa, a função inversa de f(x) é y = ln((e^2 -1)x+1)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 9999999
x = np.random.rand(N)
X = np.arange(0, 2, 0.01)
```

```python
def original_function(x):
    return np.exp(X)/(np.exp(2)-1)

def inverse_function(x):
    return np.log((np.exp(2) - 1) * x + 1)
```

```python
plt.plot(X, original_function(X))
plt.hist(inverse_function(x), bins=50, density=True)
plt.show()
```
