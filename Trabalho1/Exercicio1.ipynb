## 1) Mostre, usando análise e simulação, que o gerador de números aleatórios definido por Xi+1 = 5xi mod (7) é um gerador de período completo. Determine a sequência gerada para sementes x0 = 4 e x0 = 7. Compare as sequências e comente os resultados.


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gerador_multiplicativo(x_0, a, m, n_samples = 60):
    x_1 = np.array([x_0])
    for _ in range(n_samples):
        x_0 = (a * x_0)%m
        x_1 =  np.append(x_1,x_0)
    ind = np.arange(n_samples + 1)
    plt.bar(ind, x_1)
    plt.xlabel('amostra')
    plt.ylabel('X(i)')
    plt.show()
```

#### Para x0 = 4

```python
gerador_multiplicativo(4,5,7)
```

#### Para x0 = 7

```python
gerador_multiplicativo(7,5,7)
```

#### Um gerador de período completo é aquele que gera todas as possíveis sequências de números dentro do intervalo permitido. Um gerador de números aleatórios é considerado de período completo se tiver um período igual ao número total de valores possíveis, que neste caso seria 7. Desta forma, dados os gráficos plotados acima, podemos afimar que o gerador em questão não é um gerador de período completo.
