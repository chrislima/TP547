# 8) Utilizando o método da aceitação/rejeição gerar amostras para a distribuição f(x)=1.5x^2,−1<x<1. Plotar a pdf analitica e o histograma normalizado

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 100000
x = np.array([])
```

```python
def function(x):
    return 1.5 * x ** 2

def generate_intervals():
    x1 = np.random.uniform(-1, 1)
    x2 = np.random.uniform(0, 1.5)  
    return x1,x2
```

```python
for _ in range(N):
    x1,x2 = generate_intervals()
    while x2 >= function(x1):    
        x1,x2 = generate_intervals()
    x = np.append(x, x1)
```

```python
X=np.arange(-1, 1, 0.01)
plt.plot(X, function(X))
plt.hist(x, bins=100, density=True)
plt.show()
```
